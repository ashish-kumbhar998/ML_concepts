```python
# create dataframe with random numbers 

df = pd.DataFrame(np.random.randn(13, 4), columns=['A', 'B', 'C', 'D'], index=pd.date_range('1/1/2011', periods=13, freq='D'))
print(df)
print('---------------------------------------------------------------')
print(df.head())     # to get first 5 records
print('---------------------------------------------------------------')
print(df.tail())     # to get last 5 records
print('---------------------------------------------------------------')
print(df.index)      # print index values
print('---------------------------------------------------------------')
print(df.columns)    # print columns
print('---------------------------------------------------------------')
print(df.describe()) # describe data
```

```python
# create Dataframes by different options

df1 = pd.DataFrame({
    'A':[1,2,3,4,5],
    'B':pd.Timestamp('20220101'),
    'c':pd.Series(5, index=list(range(5)), dtype='float32'),
    'D':np.array([5]*5, dtype='int'),
    'E':pd.Categorical(['true','false','true','true','false'])
})
print(df1)
print('---------------------------------------------------------------')
print(df1.dtypes)                    # print data type
print('---------------------------------------------------------------')
print(df1.describe())                # describes the data
```

```python
# CRUD for pandas series

sr1 = pd.Series([2, 4, 6, 8, 10])
sr2 = pd.Series([1, 3, 5, 7, 9])

add = sr1+sr2
sub = sr1-sr2
mult = sr1*sr2
div = sr1/sr2

print("add",add)
print("sub",sub)
print("mult",mult)
print("div",div)
```

```python
# Convert Dictonary to pandas Series

d1 = {'a': 100, 'b': 200, 'c':300, 'd':400, 'e':800}
print("Original dictionary:")
print(d1)
new_series = pd.Series(d1)
print("Converted series:")
print(new_series)
```
